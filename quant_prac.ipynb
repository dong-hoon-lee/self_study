```python
class Ticket:
    def __init__(self, price, number):
        self.number = number
        self.price = price

    def getNumber(self):
        return self.number
    
    def getPrice(self):
        return self.price

    def setPrice(self, price):
        self.price = price

    def __str__(self):
        result = f'입장권 번호: {self.number}\n가격: {self.price}'
        return result
```

```python
class WalkupTicket(Ticket):
    def __init__(self, payCash, number, price):
        super().__init__(price, number)   # 상속받을 때는 부모 클래스를 의미하는 super() 사용하고, 어떤 인스턴스를 상속받을지 __init__ 인자로 결정
        self.payCash = payCash

    def getPrice(self):
        return self.price

    def getPayCash(self):
        if self.payCash == 'True':
            price = int(self.price * 1.05)
        else:
            price = self.price
        return price

    def __str__(self):
        result = f'현금결재: {self.payCash}\n가격: {self.price}\n번호: {self.number}'
        return result
```

```python
class AdvanceTicket(Ticket):
    def __init__(self, price, number, advanceDays):
        super().__init__(price, number)
        self.advanceDays = advanceDays

    def getPrice(self):
        return self.price

    def getAdvanceDays(self):
        days = self.advanceDays

        if days <= 4:
            rate = 0
        elif days <= 9:
            rate = 0.1
        elif days <= 29:
            rate = 0.2
        else:
            rate = 0.5

        price = int(self.price * (1-rate))
        return price

    def __str__(self):
        result = f'번호: {self.number}\n사전 예약일수: {self.advanceDays}\n가격: {self.price}'
        return result
```
