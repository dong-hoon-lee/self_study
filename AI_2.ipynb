```python
import numpy as np
import matplotlib.pyplot as plt
import sklearn
```

```python
plt.plot([1,2,3], [3,2,1])
plt.scatter([1,2,3], [3,2,1])
plt.show()
```

```python
def draw_plot(x, y, x_sample, y_sample):
    for i in range(len(x)):
        plt.plot(x[i], y[i])
    plt.scatter(x_sample, y_sample)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.show()
```

```python
foo = lambda x: -(2/7*x**3 - 9/2*x**2 + 15*x-10.)
x_line = np.linspace(0, 10, 100)
y_line = foo(x_line)
plt.plot(x_line, y_line, 'r')
plt.show()
```

```python
plt.plot(x_line)
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(x_line[:, None], y_line)

r2 = lr.score(x_line[:, None], y_line)
print(f'R^2 = {r2:.2f}')

y_hat = lr.predict(x_line[[0], None])
print(x_line[[0]])
print(y_hat)
print(f'MSE = {((y_line - y_hat)**2).mean()}')
```

```python
from sklearn.preprocessing import PolynomialFeatures

poly = PolynomialFeatures(degree=2)

x_sample_poly = poly.fit_transform(x_line[:, None])
# print(x_line[:, None])
# print(x_sample_poly)

lr = LinearRegression().fit(x_sample_poly, y_line)
y_poly = lr.predict(x_sample_poly)

draw_plot([x_line, x_line], [y_line, y_poly], x_line, y_line)
```

```python
poly = PolynomialFeatures(degree=3)

x_sample_poly = poly.fit_transform(x_line[:, None])
# print(x_line[:, None])
# print(x_sample_poly)

lr = LinearRegression().fit(x_sample_poly, y_line)
y_poly = lr.predict(x_sample_poly)

draw_plot([x_line, x_line], [y_line, y_poly], x_line, y_line)
```

```python
from sklearn.linear_model import Ridge

poly = PolynomialFeatures(degree=2)

x_sample_poly = poly.fit_transform(x_line[:, None])
# print(x_line[:, None])
# print(x_sample_poly)

lr = LinearRegression().fit(x_sample_poly, y_line)
rr = Ridge(alpha=0.9).fit(x_sample_poly, y_line)
y_poly = rr.predict(x_sample_poly)

draw_plot([x_line, x_line], [y_line, y_poly], x_line, y_line)
```
